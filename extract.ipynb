```python
url = "https://en.wikipedia.org/wiki/List_of_largest_universities"
response = requests.get(url)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.content)
```

```python
soup
```

```python
# extraction de la table ayant pour class=sortable
table = soup.find('table', attrs={'class':'sortable'})
```

```python
# selection des la premiere ligne du tableau qui contient les en-tetes de colonnes
trs = table.find_all('tr')
trs
```

```python
# selection des en-tetes de colonnes
trs[0]
```

```python
# selections des colonnes (titres)
columns = [] # liste vide
trs1= trs[0].find_all('th') 
for item in trs[0].find_all('th'): # boucle for pour extraire 
    print(item.text.strip())
    columns.append(item.text.strip())

# displaying
columns
```

```python
# we can also use this map method instead of for loop method to extract  the columns
columns = list(map(lambda x: x.text.strip(), trs[0].find_all('th')))

# rename the last column name Ref to Link
columns[-1] = 'Link'
```

```python
columns
```

```python
# extraction des lignes
rows = trs[1:]
rows
```

```python
list(map(lambda x: x.text.strip(), rows[0].find_all('td')))
```

```python
# extraction du lien de chaque university
rows[0].findAll('td')[1].a.attrs['href'].lstrip('/')
```

```python
# cette fonction permet d'extraire les donnee de chaque ligne 
def extract_row(tr):
    row_soup_list = tr.findAll('td')# extraction de toutes les cellules de la ligne
    row = list(map(lambda x: x.text.strip(), row_soup_list)) # on les met dans une liste avec la methode map
    if len(row) < 6:
        row.append("")
        print("==========================================")
        print(row)
    link = row_soup_list[1].a.attrs['href'].lstrip('/') # on extrait le lien de chaque university se trouvant dans la 2e cellule (le nom de l'university a le lien)
    row[-1] = f"https://en.wikipedia.org/{link}" if link else "" # on met le link dans la colonne link qui est la derniere colonne
    #print(f"Link : \t {link}")
    return row
    
```

```python
#extract_row(rows[30])
```

```python
# creation d'une liste de lignes  en utilisant la methode map avec la fonction extract_row()
data = list(map(extract_row, rows))
```

```python
#rows.find_all('td')
```

```python
data
```

```python
columns
```

```python
# chargement des donnees extraites dans un fichier csv
with open('universities.csv', 'w') as f:
    csvwriter = csv.writer(f)
    csvwriter.writerow(columns)
    csvwriter.writerows(data)
```

```python
with open('universities.json', 'w') as f:
    json.dump(data, f, indent=4)
```
